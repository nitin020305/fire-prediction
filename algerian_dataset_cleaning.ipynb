### Algerian Forest Fires Dataset



### 🌲 Dataset Overview

- **Name**: Algerian Forest Fires Dataset
- **Source**: UCI Machine Learning Repository & IEEE DataPort
- **Instances**: 244 total (122 from Bejaia region, 122 from Sidi Bel-Abbes)
- **Time Period**: June to September 2012
- **Purpose**: To study and predict forest fire occurrences using meteorological and fire weather index (FWI) data

---

### 📊 Attribute Information

The dataset includes **12 attributes** (features) plus a **target class**:

| Attribute | Description | Value Range |
|----------|-------------|-------------|
| `Date` | Day/Month/Year | June–Sept 2012 |
| `Temperature` | Max daily temperature (°C) | 22 to 42 |
| `RH` | Relative Humidity (%) | 21 to 90 |
| `Ws` | Wind speed (km/h) | 6 to 29 |
| `Rain` | Rainfall (mm) | 0 to 16.8 |
| `FFMC` | Fine Fuel Moisture Code | 28.6 to 94.3 |
| `DMC` | Duff Moisture Code | 0.7 to 65.9 |
| `DC` | Drought Code | 7 to 220.4 |
| `ISI` | Initial Spread Index | 0.1 to 19 |
| `BUI` | Buildup Index | 1.1 to 68 |
| `FWI` | Fire Weather Index | 0 to 31.1 |
| `Region` | Bejaia (1) or Sidi Bel-Abbes (2) | Categorical |
| `Classes` | Fire occurrence | `fire` or `not fire` |

---

### 🔍 Notes for Analysis

- The **FWI components** (FFMC, DMC, DC, ISI, BUI, FWI) are derived from the Canadian Fire Weather Index system.
- The **target variable** is `Classes`, which is binary: `fire` (1) or `not fire` (0).
- The dataset is ideal for **classification tasks**, especially using logistic regression, decision trees, or ensemble methods.

The dataset is converted into two sets based on Region from 122th index, we can make a new coloumn based on the region

1: "Bejaia Region Dataset"

2: "Sidi-Bel Abbes Region Dataset"

Add anew column with region

```python
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
df.info()
```

```python
df[['Region']]= df[['Region']].astype('int')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
## fix spaces in column names
#---------------------------------------------------------------------------------------
df.columns= df.columns.str.strip()
```
